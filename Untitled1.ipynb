```
y_label = X_train[['h1n1_vaccine','seasonal_vaccine']].values
X_train = X_train.drop(['h1n1_vaccine','seasonal_vaccine'], axis = 1).values

y_test = X_test['respondent_id'].values
X_test = X_test.drop('respondent_id', axis = 1)
X_test = X_test.drop('employment_occupation', axis = 1)
X_test = X_test.drop('employment_industry', axis = 1)
X_test = X_test.drop('census_msa', axis = 1)
X_test = X_test.drop('hhs_geo_region', axis = 1)
```

```

imp = SimpleImputer(missing_values=np.nan, strategy='mean')
X_train = imp.fit_transform(X_train)
X_test = imp.fit_transform(X_test)
```

```
#Normalize feature wise
X_train = normalize(X_train, axis = 0)
X_test = normalize(X_test, axis = 0)
```

```
def dense_block(x_input):
  
  x = tf.keras.layers.BatchNormalization()(x_input)
  x = tf.keras.layers.Dense(128, activation= 'relu')(x)
  x = tf.keras.layers.Dropout(.2)(x)
  x = tf.keras.layers.BatchNormalization()(x)
  x = tf.keras.layers.Dense(128, activation= 'relu')(x)
  x = tf.keras.layers.Dropout(.2)(x)
  x = tf.keras.layers.BatchNormalization()(x)
  x = tf.keras.layers.Dense(128, activation= 'relu')(x)
  x = tf.keras.layers.Dropout(.2)(x)
  x = tf.keras.layers.BatchNormalization()(x)

  return x

def fusion_block(value_1, value_2, previous_fusion):


  #added = tf.keras.layers.Add()([value_1, value_2])
  #Value 1 is a relict, which was used in ythe past
  dense = tf.keras.layers.concatenate([previous_fusion, value_2])
  return dense
```

```
def scheduler(epoch, lr):
  if epoch < 15:
    lr = 0.001
    return lr
  elif epoch > 15 and epoch < 25:
    lr = 0.0001
    return lr
  else:
    lr = 0.00001
    return lr
callback = tf.keras.callbacks.LearningRateScheduler(scheduler)
```

```
import tensorflow as tf
```

```
tf.keras.backend.clear_session()
x_input = tf.keras.Input(shape=(31,))

x_b = tf.keras.layers.Dense(128, activation= 'relu')(x_input)

x = dense_block(x_b)
f = fusion_block(x_b, x, x_b)

x1 = dense_block(f)
f1 = fusion_block(x, x1, f)

x2 = dense_block(f1)
f2 = fusion_block(x1, x2, f1)

x3 = dense_block(f2)
f3 = fusion_block(x2, x3, f2)

x4 = dense_block(f3)
f4 = fusion_block(x3, x4, f3)

x5 = dense_block(f4)
f5 = fusion_block(x4, x5, f4)

x6 = dense_block(f5)
f6 = fusion_block(x5, x6, f5)

x7 = dense_block(f6)
f7 = fusion_block(x6, x7, f6)

x8 = dense_block(f7)
f8 = fusion_block(x6, x7, f7)

x9 = dense_block(f8)
f9 = fusion_block(x7, x8, f8)

x10 = dense_block(f9)
f10 = fusion_block(x8, x9, f9)

x11 = dense_block(f10)
f11 = fusion_block(x9, x10, f10)

x12 = dense_block(f11)
f12 = fusion_block(x10, x11, f11)

x13 = dense_block(f12)
f13 = fusion_block(x11, x12, f12)

x14 = dense_block(f13)
f14 = fusion_block(x12, x13, f13)

x15 = dense_block(f14)
f15 = fusion_block(x13, x14, f14)

x16 = dense_block(f15)
f16 = fusion_block(x14, x15, f15)

x17 = dense_block(f16)
f17 = fusion_block(x15, x16, f16)

x18 = dense_block(f17)
f18 = fusion_block(x16, x17, f17)

x19 = dense_block(f18)
f19 = fusion_block(x17, x18, f18)

x20 = dense_block(f19)
f20 = fusion_block(x18, x19, f19)


f20 = tf.keras.layers.Dense(512, activation= 'relu')(f20)
f20 = tf.keras.layers.Dropout(.2)(f20)
f20 = tf.keras.layers.Dense(512, activation= 'relu')(f20)
f20 = tf.keras.layers.Dropout(.2)(f20)

output = tf.keras.layers.Dense(2, activation="sigmoid")(f20)

model = tf.keras.Model(x_input, output)
model.compile(optimizer='Adam',
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=[tf.keras.metrics.AUC()])
```

```
model.summary()
```

```
light_fire3 = model.fit(X_train, y_label, epochs=40, batch_size=64, validation_split= 0.2,
              validation_data= (X_train, y_label), callbacks=[callback])
```

```
plt.plot(light_fire3.history['val_auc'], label = 'Val Auc')
plt.plot(light_fire3.history['auc'], label = 'Train Auc')
plt.legend()
```
